# 2. Missing Value Analysis

<b>a. Search The Missing Value

```python
missing_data        = df.isna().sum()
missing_percentage  = round((df.isna().sum()/len(df)*100), 2)

missing_value = pd.DataFrame({
    'Missing Data'  : missing_data,
    'Percentage(%)' : missing_percentage
})

missing_value
```

Note:<br>
The data shows that 'Income' has 24 missing values, which is 1.07% from total dataset. <br>
In general, 1% missing values do not have a significant impact on the analysis. However, further checks will still be conducted to handle these missing values.

<b>b. Check The Distribution of Data</b><br>
<br>(1) Check the distribution of data (normal/not) in 24 missing values in 'Income' 
<br>(2) Next, the missing value will be filled with: (a) <i>Mean</i>, if the distribution of data is normal; (b) <i>Median</i>, if the distribution of data is not normal

```python
# Check the data distribution (normal/not)

from scipy.stats import shapiro
from scipy.stats import normaltest

income_data = df['Income'].dropna()

shapiro_stat, shapiro_income = shapiro(income_data)
if shapiro_income > 0.05 :
    print(f'For the missing value in income data:\nWe assume as a normal distribution, because P-Value result of Shapiro Wilk Test ({shapiro_income}) is larger than 0.05.')
    print(f'In this case, the missing values can be replaced with the mean value of {df['Income'].mean()}.')
else :
    print(f"For the missing value in income data:\nWe don't assume as a normal distribution, because P-Value result of Shapiro Wilk Test ({shapiro_income}) is smaller or equal to 0.05.")
    print(f'In this case, the missing values can be replaced with the median value of {df['Income'].median()}.')
```

<b>c. Check Outliers</b><br>

Before filling in the missing value,vwe will check first for the outliers in the selected numerical data such as:<br>
* Product   : 'Wine', 'Fruit', 'Meat', 'Fish', 'Sweet', 'Gold'<br>
* Place     : 'Total_Purch+Disc', 'Web_Purchase', 'Catalog_Purchase', 'Store_Purchase', 'Web_Visit'<br> 
* People    : 'Income'

```python
plt.figure(figsize=(20, 10))

# List of numerical data for further analysis
numerical_df = ['Wine', 'Fruit', 'Meat', 'Fish', 'Sweet', 'Gold',
                'Total_Purch+Disc', 'Web_Purchase', 'Catalog_Purchase', 'Store_Purchase', 'Web_Visit', 
                'Income']

# Create boxplots for each numerical feature
for i, num in enumerate(numerical_df, 1):
    plt.subplot(4, 3, i)  # Adjust the number of rows/columns as needed
    sns.boxplot(x=df[num])
    plt.title(num)

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import normaltest, skew
plt.figure(figsize=(15, 10))

for i, col in enumerate(numerical_df):
    plt.subplot(4, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')

    # Normality test
    k2, p = normaltest(df[col])
    if p > 0.05 :
        normality = 'Normal'
    else: 
        normality = 'Not Normal'

    # Skewness
    skewness = skew(df[col])
    if skewness > 0   :
        skew_type = 'Positive Skew'
    elif skewness < 0 : 
        skew_type = 'No Skew'
    else: 
        skew_type = 'Negative Skew'

    plt.text(0.95, 0.95, f"Normality: {normality}\nSkew: {skew_type} ({round(skewness,2)})",
             transform=plt.gca().transAxes, ha='right', va='top')

plt.tight_layout()
plt.show()
```

Note:<br>
1. The results of the boxplot and histogram shows not normal distribution of selected numerical data such as: ['Wine', 'Fruit', 'Meat', 'Fish', 'Sweet', 'Gold', 'Total_Purch+Disc', 'Web_Purchase', 'Catalog_Purchase', 'Store_Purchase', 'Web_Visit', 'Income']. <br>
2. 'Income' column has outliers that are significantly distant from the data range (around 600,000). This outlier will be removed. <br>
3. Outliers in the other data will be ignored, because there are assumption that customer purchases may happen under various conditions (example: different purchases on regular days vs on holidays).

Remove 'Income' Outlier

```python
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Income'] >= lower_bound) & (df['Income'] <= upper_bound)]

# Histogram and boxplot of 'Income' after outlier removal
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# Histogram
sns.histplot(df['Income'], kde=True, bins=50, ax=ax[0], color='blue')
ax[0].set_title("'Income' Histogram\n(after outlier removal)", size=20)
ax[0].set_xlabel('Income')
ax[0].set_ylabel('Frequency')

# Boxplot
sns.boxplot(x=df['Income'], ax=ax[1], color='lightblue')
ax[1].set_title("'Income' Boxplot\n(after outlier removal)", size=20)
ax[1].set_xlabel('Income')

plt.show()
```

Insert new data for missing values (based on the statistical results of the data distribution)

```python
income_median = df['Income'].median()
df.loc[:, 'Income']  = df['Income'].fillna(income_median)

df.isna().sum()
```

<b>(1) 'Year_Birth' classification</b><br>
* 'Year_Birth' will be classified into new column as 'Group_Generation' (consist of: Baby Boomer, Gen X, Millenial, Gen Z, Alpha)<br>
* Data will be used to analyze the effect of generation classification on the purchase of goods.

```python
df['Year_Birth'].value_counts().sort_index(ascending=False)
```

```python
def categorize_bygeneration(year) :      # based on kemenkeu RI
    if year <= 1964             :
        return 'Baby Boomer (<=1964)'
    elif 1965 <= year <= 1980   :
        return 'Gen X (1965-1980)'
    elif 1981 <= year <= 1996   :
        return 'Millennial (1981-1996)'
    elif 1997 <= year <= 2012   :
        return 'Gen Z (1997-2012)'
    elif year >= 2013           :
        return 'Alpha (>=2013)'
    else:
        return 'Unknown'

df['Group_Generation'] = df['Year_Birth'].apply(categorize_bygeneration)
count_generation       = df['Group_Generation'].value_counts()
count_generation
```

```python
plt.figure(figsize=(8, 5))
plt.pie(count_generation, labels=count_generation.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Customer Distribution\n(based on group generation)', size=18)
plt.show()
```

<b>(2) 'Education' classification</b> <br>
* 'Education' will be grouped into 3 categories: postgraduate, graduate, and undergraduate.<br>
* Data will be used to analyze the effect of education on the purchase of goods.

```python
df['Education'].value_counts()
```

```python
def categorize_edu(edu):
    if edu in ['PhD', 'Master']  :
        return 'Postgraduate'
    elif edu in ['Graduation']   :
        return 'Graduate'
    elif edu in ['2n Cycle', 'Basic'] :
        return 'Undergraduate'
    else:
        return 'Unknown'

df['Education'] = df['Education'].apply(categorize_edu)
df['Education'].value_counts()
```

```python
count_education = df['Education'].value_counts()

plt.figure(figsize=(8, 5))
plt.pie(count_education, labels=count_education.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Customer Distribution\n(based on education)', size=18)
plt.show()
```

### <b>Campaign Further Analysis</b>

    'TotalAcc_Campaign'
* Total acceptance of campaign offers: Acc1_Camp, Acc2_Camp, Acc3_Camp, Acc4_Camp, Acc5_Camp, LastAcc_Camp
<br>Note:<br>
    * Acc1_Camp     : 1 if the customer accepted the offer in the 1st campaign, 0 otherwise<br>
    * Acc2_Camp     : 1 if the customer accepted the offer in the 2nd campaign, 0 otherwise<br>
    * Acc3_Camp     : 1 if the customer accepted the offer in the 3rd campaign, 0 otherwise<br>
    * Acc4_Camp     : 1 if the customer accepted the offer in the 4th campaign, 0 otherwise<br>
    * Acc5_Camp     : 1 if the customer accepted the offer in the 5th campaign, 0 otherwise<br>
    * LastAcc_Camp  : 1 if the customer accepted the offer in the last campaign, 0 otherwise<br>

```python
df['Total_Accepted'] = df[['Acc1_Camp', 'Acc2_Camp', 'Acc3_Camp', 'Acc4_Camp', 'Acc5_Camp','LastAcc_Camp']].sum(axis=1)
df['Cust_AccAll']    = (df['Total_Accepted'] == 6).astype(int)

target_cust = df[df['Cust_AccAll'] == 1]
print(f"Total customer(s) who accepted all the campaigns: {target_cust.shape[0]}")
```

Based on the results, there are no customers who accepted all the campaigns.<br> 
To determine the target customer, we will analyze customers who accepted the total campaigns in sequence, from the first to the last.

```python
# Customer acceptance criteria
def acc_criteria(row):
    # cust acc all campaign
    if row['Acc1_Camp'] == 1 and row['Acc2_Camp'] == 1 and row['Acc3_Camp'] == 1 and row['Acc4_Camp'] == 1 and row['Acc5_Camp'] == 1 and row['LastAcc_Camp'] == 1:
        return 6  
    # cust acc 5 campaign
    elif row['Acc1_Camp'] == 1 and row['Acc2_Camp'] == 1 and row['Acc3_Camp'] == 1 and row['Acc4_Camp'] == 1 and row['Acc5_Camp'] == 1 and row['LastAcc_Camp'] == 0:
        return 5  
    # cust acc 4 campaign
    elif row['Acc1_Camp'] == 1 and row['Acc2_Camp'] == 1 and row['Acc3_Camp'] == 1 and row['Acc4_Camp'] == 1 and row['Acc5_Camp'] == 0:
        return 4  
    # cust acc 3 campaign
    elif row['Acc1_Camp'] == 1 and row['Acc2_Camp'] == 1 and row['Acc3_Camp'] == 1 and row['Acc4_Camp'] == 0:
        return 3  
    # cust acc 2 campaign
    elif row['Acc1_Camp'] == 1 and row['Acc2_Camp'] == 1 and row['Acc3_Camp'] == 0:
        return 2  
    # cust acc 1 campaign
    elif row['Acc1_Camp'] == 1 and row['Acc2_Camp'] == 0:
        return 1  
    # cust not accept all
    elif row['Acc1_Camp'] == 0:
        return 0  

df['acc_criteria'] = df.apply(acc_criteria, axis=1)
all_target = df[df['acc_criteria'].isin([0, 1, 2, 3, 4, 5, 6])]

# Total customer based on customer acceptance criteria (acc_criteria)
summary = all_target['acc_criteria'].value_counts().sort_index()

summary_df = pd.DataFrame(summary).reset_index()
summary_df.columns = ['Total_Campaign', 'Total_Customer']

# Fill the campaign 0 - 6
summary_df = summary_df.set_index('Total_Campaign').reindex(range(7), fill_value=0).reset_index()
summary_df
```

Note:<br>
The data shows that customers typically accept up to three campaigns, after which the response rate drops to zero. This suggests that campaigns after the third one tend to lose their effectiveness, possibly due to factors such as lack of relevance or decreased of interest. It indicates that after a certain point, customers may no longer find the campaigns appealing, highlighting the need for more varied or personalized offers to maintain engagement.

```python
plt.figure(figsize=(10, 6))
plt.bar(summary_df['Total_Campaign'].astype(str), summary_df['Total_Customer'], color='steelblue')
plt.title('Total Customer Based on Campaign Acceptance')
plt.xlabel('Total Campaign Accepted')
plt.ylabel('Total Customer')
plt.xticks()
plt.show()
```

1. <b>Target Market</b>

* Total Family variable has a positive correlation with discount-based purchases, suggesting that families are responsive to promotions and discounts. Creating family-oriented discounts for products that commonly purchased by families (like groceries and household essentials) would attract this market.<br>

<br><br>

3. <b> Seasonal Campaigns</b>
* Seasonal events such as holiday and personal milestones can increase engagement and sales across multiple product categories. Using this opportunity to create limited-time campaigns for specific products can create sense of urgency and increase the engagement.<br>
* Reward loyalty program that offers points based on campaign participation can enhance loyalty and make customers more likely to engage with future promotions.<br>
